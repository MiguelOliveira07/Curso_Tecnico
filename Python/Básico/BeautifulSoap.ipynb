```python
from bs4 import BeautifulSoup
import requests

url = 'https://www.dadosmundiais.com/moedas/'

resposta = requests.get(url, verify = False)
sopa = BeautifulSoup(resposta.text, 'html.parser')

sopa.prettify()
```

```python
moedas = sopa.find_all('td')

for moeda in moedas:
    print(moeda.text)
    print('-----')
```

```python
lista_links = []

links = sopa.find_all('a')

for link in links:
    if 'https://' in link.get('href'):
        lista_links.append(link.get('href'))

print(lista_links)
```
